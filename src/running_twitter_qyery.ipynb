```python
days = 115
primeira_lista_de_datas = []
while days > 0:
    data = (datetime.today() - timedelta(days=days)).strftime("%Y-%m-%d")
    days -= 6
    primeira_lista_de_datas.append(data)

days = 115
segunda_lista_de_datas = []
while days > 0:
    days -= 6
    data = (datetime.today() - timedelta(days=days)).strftime("%Y-%m-%d")
    segunda_lista_de_datas.append(data)

df = pd.DataFrame(data={'primeira_data': primeira_lista_de_datas, 'segunda_data': segunda_lista_de_datas})
```
